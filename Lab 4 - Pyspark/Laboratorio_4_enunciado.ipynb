````python
#5. Cree una funci√≥n cast_columns que permita cambiar el tipo de datos de las columnas problem√°ticas.
# Luego utilice esta funci√≥n respecto a lo respondido en la pregunta anterior.
def cast_columns(df, cols_types):
    """
    Cambia el tipo de m√∫ltiples columnas en un DataFrame de Spark.

    Par√°metros:
    - df: DataFrame de Spark.
    - cols_types: Diccionario con nombres de columnas como claves y tipos de datos de Spark como valores.

    Retorna:
    - DataFrame de Spark con tipos de columnas modificados.
    """
    for col, new_type in cols_types.items():
        df = df.withColumn(col, df[col].cast(new_type))
    return df
````

````python
'''
Con la funci√≥n realizada se modifican las columnas:
-Is An Alien de Double a Boolean, ya que son datos de 0 a 1 para indicar si es o no, lo que se puede transformar en True o False sin problema
-Account Age Days, Transaction Hour, Quantity y Customer Age se pasan de Double a Integer ya que son valores que no presentan ni presentar√≠an decimales.
'''
dict_types = {
    'Transaction ID': 'string',
    'Customer ID': 'string',
    'Transaction Amount': 'double',
    'Transaction Date': 'timestamp_ntz',
    'Payment Method': 'string',
    'Product Category': 'string',
    'Quantity': 'integer',
    'Customer Age': 'integer',
    'Customer Location': 'string',
    'Device Used': 'string',
    'IP Address': 'string',
    'Shipping Address': 'string',
    'Billing Address': 'string',
    'Is An Alien': 'boolean',
    'Account Age Days': 'integer',
    'Transaction Hour': 'integer'
}
df = cast_columns(df, dict_types)
df.show(10)
df.dtypes
````

### 3. Transformaciones con pyspark [6 puntos]
(1 punto por pregunta)

**Nota: NO SE PERMITE EL USO DE PANDAS EN ESTA SECCI√ìN**

Para continuar con el an√°lisis, los especistas de Bodoque les gustar√≠a tener nuevas variables disponibles. Tras las notas de la reuni√≥n usted llega a la conclusi√≥n de que tiene que realizar las siguientes tareas (con el dataset preprocesado de la seccion anterior):


1.   Agregar una columna llamada "Transaction bp" con el **monto total** de la transacci√≥n en bodoque pesos. Se considera que $x$ d√≥lares equivalen a $log(48+|x^{36}|)$ bodoque pesos.
2.   Crear una columna llamada "Transaction Month" con el mes en que se realiza una transacci√≥n.
3.   Crear la variable *Type of purchase* seg√∫n la catidad de unidades vendidas de acuerdo a las siguientes categor√≠as.
  * Compra minorista: 4 productos o menos.
  * Compra mayorista: 5 produtos o m√°s.
4. Imprima los registros de compras hechas por alien√≠genas en el comecio mayorista.  Utilice `.filter()`.
5. Cuente la cantidad de compras realizadas por humanos y la cantidad de compras realizadas por alien√≠genas. Utilice `.groupby()`.
6. Muestre una tabla con la recaudaci√≥n promedio por transacci√≥n para cada m√©todo de pago, tanto para humanos como alien√≠genas. Utilice `pivot()`.

````python
from pyspark.sql.functions import col, log, month, mean, when, count, round
#1.- Agregar una columna llamada "Transaction bp" con el monto total de la transacci√≥n en bodoque pesos.
print("1.- Transaction bp")
df = df.withColumn("Transaction_bp", log(48 + (col("Transaction Amount") * 36)))
df.select("Transaction Amount", "Transaction_bp").show()

#2.- Crear una columna llamada "Transaction Month" con el mes en que se realiza una transacci√≥n.
print("2.- Transaction Month")
df = df.withColumn("Transaction_Month", month("Transaction Date"))
df.select("Transaction Date", "Transaction_Month").show()

#3.- Crear la variable Type of purchase seg√∫n la catidad de unidades vendidas
print("3.- Type of purchase")
df = df.withColumn("Type_of_purchase", when(col("Quantity") <= 5, "Compra minorista").otherwise("Compra mayorista"))
df.select("Type_of_purchase").show()

#4.- Imprima los registros de compras hechas por alien√≠genas en el comecio mayorista.
print("4.-  Registro Compras Alienigenas Mayoristas")
df.filter((col("Is An Alien") == True) & (col("Type_of_purchase") == "Compra mayorista")).show()

#5.- Cuente la cantidad de compras realizadas por humanos y la cantidad de compras realizadas por alien√≠genas.
print("5.-  Compras humanas y Alienigenas")
df.groupby("Is An Alien").agg(count("*").alias("Cantidad_de_compras")).show()

#6.- Muestre una tabla con la recaudaci√≥n promedio por transacci√≥n para cada m√©todo de pago, tanto para humanos como alien√≠genas.
print("6.- Tabla Recaudaci√≥n")
recaudacion = df.groupby("Payment Method", "Is An Alien").agg(mean("Transaction Amount").alias("Recaudacion_promedio"))
recaudacionpromedio = recaudacion.groupBy("Payment Method").pivot("Is An Alien").agg(round(mean("Recaudacion_promedio"), 2))
recaudacionpromedio.show()
````

Las funciones de ventana permiten realizar c√°lculos sobre un conjunto de filas que est√°n relacionadas con la fila actual. Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la funci√≥n de ventana comienza desde la primera fila de la partici√≥n o del conjunto de resultados, lo cual es √∫til para calcular sumas acumulativas hasta la fila actual. Las variaciones comunes de este uso incluyen:

- `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partici√≥n hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partici√≥n.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango espec√≠fico basado en valores antes y despu√©s de la fila actual.

````python
# 1.- Cargar el objeto serializado: Revise el tipo de objeto y deduzca su funci√≥n.
import pickle
with open(object_path, 'rb') as file:
    object_pkl = pickle.load(file)
print(type(object_pkl))
````

El objeto corresponde a un modelo de regresi√≥n log√≠stica entrenado (LogisticRegression)

````python
# 2.- Explorar el objeto: Utilice las funciones dir y help para identificar qu√© m√©todo del objeto predice la probabilidad.
print(dir(object_pkl))
help(object_pkl)
````

````python
# 3.- Crear una funci√≥n personalizada: Elabore una funci√≥n que prediga la probabilidad de fraude
#utilizando el √∫ltimo valor de la lista generada por el objeto serializado. Puede modificar el nombre de la funci√≥n para reflejar su prop√≥sito.
def es_fraude(transaction_amount, quantity, customer_age, transaction_hour):
  #Ya que nos piden predecir probabilidad utilizaremos la funcion predict_proba existente en el pkl
  predicted_probability = object_pkl.predict_proba([[transaction_amount, quantity, customer_age, transaction_hour]])
  #Posteriormente se retorna el √∫ltimo valor del array y nos aseguramos que sea de tipo Float
  return float(predicted_probability[-1][-1])
````

````python
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType
# 4.- Definir la funci√≥n UDF: Establezca la funci√≥n UDF con la funci√≥n personalizada que ha creado.
#Utilizando las instrucciones anteriores y nuestra funci√≥n
udf_fraude = udf(es_fraude, FloatType())
````

````python
# 5.- Generar una nueva columna: A√±ada una nueva columna prediction a su DataFrame en Spark utilizando la funci√≥n UDF y muestre un ejemplo de c√≥mo se aplica.
df = df.withColumn("Fraud_prediction", udf_fraude("Transaction Amount", "Quantity", "Customer Age", "Transaction Hour"))
````
