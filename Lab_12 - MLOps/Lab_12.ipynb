## **MLFlow**

`MLflow` es una plataforma de c칩digo abierto que simplifica la gesti칩n y seguimiento de proyectos de aprendizaje autom치tico. Con sus herramientas, los desarrolladores pueden organizar, rastrear y comparar experimentos, adem치s de registrar modelos y controlar versiones.

Si bien esta plataforma cuenta con un gran n칰mero de herramientas y funcionalidades, en este laboratorio trabajaremos con dos:
1. **Runs**: Registro que constituye la informaci칩n guardada tras la ejecuci칩n de un entrenamiento. Cada `run` tiene su propio run_id, el cual sirve como identificador para el entrenamiento en s칤 mismo. Dentro de cada `run` podremos acceder a informaci칩n como los hiperpar치metros utilizados, las m칠tricas obtenidas, las librer칤as requeridas y hasta nos permite descargar el modelo entrenado.
2. **Experiments**: Se utilizan para agrupar y organizar diferentes ejecuciones de modelos (`runs`). En ese sentido, un experimento puede agrupar 1 o m치s `runs`. De esta manera, es posible tambi칠n registrar m칠tricas, par치metros y archivos (artefactos) asociados a cada experimento.

````python
#lectura base de datos
df = pd.read_csv(os.path.join(path, 'water_potability.csv'))
df
````

````python
#con los valores nlos los eliminamos
df = df.dropna()
````

````python
#separamos el dataset en la clase y el resto de columnas
x = df.drop(columns=['Potability'])
y = df['Potability']

#separo en conjunto entrenamiento, valid y test
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2, random_state=60)
````

````python
def opt_hiper(trial):
  param={
        "objective": "binary:logistic",
        "eval_metric": "logloss",
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
        "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "min_child_weight": trial.suggest_float("min_child_weight", 1, 10),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "eta": trial.suggest_float("eta", 0.01, 0.1),
        "gamma": trial.suggest_float("gamma", 0.01, 1.0)

  }

  #creamos y entrenamos el modelo
  model = xgb.XGBClassifier(**param)
  model.fit(x_train, y_train)

  #vemos las predis y se calcula f1
  y_pred = model.predict(x_test)
  valid_f1 = f1_score(y_test, y_pred, average='weighted')
  if mlflow.active_run():
    mlflow.end_run()  # Finaliza cualquier ejecuci칩n activa

  with mlflow.start_run(run_name=f"XGBoost Trial {trial.number}"):
    mlflow.log_params(param)
    mlflow.log_metric("valid_f1", valid_f1)

  return valid_f1
````

````python
def optimize_model():
  mlflow.set_experiment('lab12')

  #se crea el estudio de optuna
  study=optuna.create_study(direction="maximize")

  #se optimiza
  study.optimize(opt_hiper,n_trials=100)

  #se obtiene el mejor modelo
  best_trial=study.best_trial
  best_model=xgb.XGBClassifier(**best_trial.params)
  best_model.fit(x_train,y_train)

  #guardamos el mejorcito
  with open(os.path.join(path, 'best_model.pkl'), 'wb') as f:
    pickle.dump(best_model, f)

  mlflow.log_artifact(os.path.join(path, 'best_model.pkl'))

  #obtenemos gr치ficos y guardamos
  fig1=optuna.visualization.plot_optimization_history(study)
  fig1.write_image(os.path.join(path, 'optimization_history.png'))
  fig2=optuna.visualization.plot_param_importances(study)
  fig2.write_image(os.path.join(path, 'param_importances.png'))

  mlflow.log_artifact(os.path.join(path, 'optimization_history.png'))
  mlflow.log_artifact(os.path.join(path, 'param_importances.png'))

  return best_model
if __name__ == "__main__":
  best_model=optimize_model()

````

**Respuestas 3.2:**

1.- Docker se diferencia principalmente de las VMs a trav칠s de su funcionamiento, mientras que Docker usa contenedores que permiten ejecutar aplicaciones y sus requerimientos en un entorno espec칤fico y aislado, las VMs virtualizan una copia del hardware completo causando as칤 que sean pesadas de correr y utilizar que los contenedores de Docker.

2.- La diferencia que existe corresponde a que cuando utilizas el sistema local se pueden ocasionar problemas de compatibilidad entre aplicaciones instaldads y versiones de estas mientras que Docker corre la aplicaci칩n y sus dependencias en un contenedor, asegurando que se ejecute en un entorno espec칤fico para el funcionamiento de esta, independiente de la configuraci칩n del sistema local.

3.- Docker asegura esta consistencia utilizando "im치genes" para garantizar que la aplicaci칩n y sus dependencias sean id칠nticas en todos los entornos. Estas im치genes contienen el sistema operativo base, las dependencias de la aplicaci칩n y la configuraci칩n necesaria para ejecutarla. Luego al ejecutar este contenedor el entorno que utiliza siempre correponde al mismo independiente de en que plataforma se ejecute.

4.- Los vol칰menes se emplean en Docker para conservar los datos generados y utilizados por los contenedores. Estos pueden ser directorios montados desde el sistema host o vol칰menes gestionados directamente por Docker. Gracias a esto, los datos permanecen disponibles incluso si un contenedor se elimina. Al configurar un contenedor, es posible definir vol칰menes para garantizar que informaci칩n clave, como bases de datos o archivos de configuraci칩n, se mantenga intacta durante reinicios o actualizaciones.

5.- Dockerfile corresponde a un archivo de texto que contiene instrucciones para construir una imagen de Docker permitiendo as칤 definir c칩mo se construye el entorno de la aplicaci칩n y hacer que sea reproducible de una forma autom치tica. "docker-compose.yml" es un recurso que permite describir de manera clara y estructurada los servicios, redes y vol칰menes necesarios para una aplicaci칩n que utiliza m칰ltiples contenedores Docker. Facilita la configuraci칩n y administraci칩n de estos servicios, gestion치ndolos como una unidad integrada. Esto resulta especialmente 칰til para aplicaciones complejas, como aquellas que combinan un backend, una base de datos y un servidor de cach칠, ya que simplifica su despliegue e interconexi칩n de forma eficiente y ordenada.
